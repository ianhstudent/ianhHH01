```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 10
x1 = range(n)
x2 = range(n)

x1_train = np.array(x1, dtype=np.float32) # float32 customary in ML big data
x1_train = x1_train.reshape(-1, 1)        # Change from vector to 2-d array
c1 = -2.                      # to make it random: np.random.randn()
c2 = 1.                       # to make it random: np.random.randn()
x2_train = np.array(x2, dtype=np.float32)
x2_train = x2_train.reshape(-1, 1)
c3 = -2
y_train = c1*x2_train + c2 + c3*x2_train
y_train = np.array(y_train + np.random.randn(n,1), dtype=np.float32)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
import torch
import torch.nn as nn
from torch.autograd import Variable

def forward(x1,x2):                      # forward pass: use model to predict outputs
  return w1*x1 + b +w2*x2

def criterion(yhat,y):                    # mean square error
  return torch.mean((yhat-y)**2)

LOSS = []
lr = 0.01
w1 = torch.tensor(0., requires_grad=True) # data parameters with initial guess
w2 = torch.tensor(0., requires_grad=True)
b = torch.tensor(-0.5, requires_grad=True)
x1_traint = torch.tensor(x1_train) # convert to pytorch tensors
x2_traint = torch.tensor(x2_train)
y_traint = torch.tensor(y_train)

def train_model(iter):
    
    # Loop
    for epoch in range(iter):
        
        # make a prediction
        Yhat = forward(x1_traint,x2_traint)
        
        # calculate the loss 
        loss = criterion(Yhat, y_traint)
            
        # store the loss in the list LOSS
        LOSS.append(loss)
        
        # backward pass, find gradient of loss wrt parameters
        loss.backward()
        
        # Stochastic gradient descent: update parameters slope and bias
        w1.data = w1.data - lr * w1.grad.data
        w2.data = w2.data - lr * w2.grad.data
        b.data = b.data - lr * b.grad.data
        
        # zero the gradients before running the backward pass
        w1.grad.data.zero_()
        w2.grad.data.zero_()
        b.grad.data.zero_()

train_model(500)
```

```python
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(LOSS)
plt.show()
```

```python
class LinearRegressionModel(nn.Module):
  def __init__(self, input_dim, output_dim):
    super(LinearRegressionModel, self).__init__()
    self.linear = nn.Linear(input_dim, output_dim) # one layer

  def forward(self, x1, x2):
    predict_y = self.linear(x1, x2)
    return predict_y
```

```python
# make an instance of that class
input_dim = 1
output_dim = 1

model = LinearRegressionModel(input_dim, output_dim)
criterion = nn.MSELoss() # mean square error as loss function

learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

inputs = Variable(torch.tensor(x1_train), torch.tensor(x2_train))
labels = Variable(torch.tensor(y_train))

epochs = 50
LOSS = []
for epoch in range(epochs):
  optimizer.zero_grad() # zero out each step, don't accumulate gradients
  predict_y = model(inputs)
  loss = criterion(predict_y, labels)
  loss.backward()       # back-propagate loss to get gradients wrt parameters
  optimizer.step()      # take an optimization step

  LOSS.append(loss)
  # print('epoch {}, loss {}'.format(epoch + 1, loss.item()))
```
